```python
# Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
# Import Dataset
ds=pd.read_csv("Position_Salaries.csv")
ds
```

```python
# Splitting dataset into Independent and Dependent variables
x=ds[["Level"]]
y=ds[["Salary"]]
```

```python
x
```

```python
y
```

```python
# Fitting Random Forest Regression Model
from sklearn.ensemble import RandomForestRegressor
reg=RandomForestRegressor(n_estimators=10,random_state=0)
reg.fit(x,y)
```

```python
# Predict for myinput
myinput=pd.DataFrame(data=[[4]],columns=['Level'])
reg.predict(myinput)
#reg.predict([[4]])
```

```python
# Visualization of random forest regression tree
from sklearn.tree import plot_tree
plot_tree(reg.estimators_[0],filled=True)
plt.show()
```

```python
plot_tree(reg.estimators_[2],filled=True)
plt.show()
```

```python
# Predict for uset Input
level=float(input("Enter the Level:"))
myinput=pd.DataFrame(data=[[level]],columns=["Level"])
result=reg.predict(myinput)
print(f"Predicted Salary for Level{level}:₹{result[0]}")
```
